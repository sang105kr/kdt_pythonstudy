```python
# 필요한 라이브러리를 불러옵니다.
import pandas as pd

# 데이터를 로드합니다.
df = pd.read_csv('인사정보.csv')
```

```python
df.head(5)
```

```python
df.shape
```

```python
# case1)
print(df['부서'].value_counts())

# case2)
print(df.groupby('부서')['부서'].count().sort_values(ascending=False))
```

```python
print(df.groupby('직급')['연봉(만원)'].mean())
print(df.groupby('직급')['연봉(만원)'].max())
print(df.groupby('직급')['연봉(만원)'].min())
print(df.groupby('직급')['연봉(만원)'].std())
print(df.groupby('직급')['연봉(만원)'].var())
```

```python
df[df['입사년도'] >= 2010]
len(df[df['입사년도'] >= 2010])
df[df['입사년도'] >= 2010].shape[0]
```

```python
df[df['연봉(만원)'] >= 6000]
```

```python
df.loc[df['부서'] == '개발', '연봉(만원)'].mean()
```

```python
df.groupby('부서')['연봉(만원)'].max() # value 
df.groupby('부서')['연봉(만원)'].idxmax() # 위치인덱스
df.iloc[df.groupby('부서')['연봉(만원)'].idxmax()]
```

```python
df['입사년도'].value_counts().sort_values()
# case 1)
print(df['입사년도'].value_counts().sort_index())
# case2)
print(df.groupby('입사년도')['입사년도'].count().sort_index())
```

```python
df.groupby('직급')['연봉(만원)'].agg(['max','min','mean'])
```

```python
# case1)
df.sort_values(by='연봉(만원)',ascending=False).head(10).reset_index(drop=True)

# case2)
df.nlargest(10,'연봉(만원)').reset_index(drop=True)
```

```python
df.iloc[df.groupby('부서')['입사년도'].idxmin()]
```

```python
df.groupby('부서').apply(lambda x : x.nlargest(3,'연봉(만원)'))
```

```python
df.groupby('부서')['연봉(만원)'].agg(['max','min'])
df.groupby('부서')['연봉(만원)'].agg(lambda x:x.max()-x.min()).idxmax()
```
