# 판다스 연습 문제 모범 답안

## 문제와 답안

```python
# 필요한 라이브러리를 불러옵니다.
import pandas as pd

# 데이터를 로드합니다.
df = pd.read_csv('인사정보.csv')
```

### CSV 파일을 읽고 데이터프레임으로 로드하세요.

```python
pd.read_csv('인사정보.csv')
```

### 데이터프레임의 처음 5행을 출력하세요.

```python
df.head()
```

### 데이터프레임의 구조(행과 열의 수)를 확인하세요.

```python
df.shape
```

### 각 부서별 직원 수를 계산하세요.

```python
df['부서'].value_counts()
```

### 직급별 평균 연봉을 계산하세요.

```python
df.groupby('직급')['연봉(만원)'].mean()
```

### 2010년 이후 입사한 직원의 수를 계산하세요.

```python
df[df['입사년도'] >= 2010].shape[0]
```

### 연봉이 6000만원 이상인 직원들의 정보를 출력하세요.

```python
df[df['연봉(만원)'] >= 6000]
```

### '개발' 부서 직원들의 평균 연봉을 계산하세요.

```python
df[df['부서'] == '개발']['연봉(만원)'].mean()
```

### 각 부서별로 가장 높은 연봉을 받는 직원의 정보를 출력하세요.

```python
df.loc[df.groupby('부서')['연봉(만원)'].idxmax()]
```

### 입사년도별 직원 수를 계산하고, 그 결과를 입사년도로 정렬하세요.

```python
df['입사년도'].value_counts().sort_index()
```

### 각 직급별 연봉의 최대값, 최소값, 평균값을 계산하세요.

```python
df.groupby('직급')['연봉(만원)'].agg(['max', 'min', 'mean'])
```

### 연봉 상위 10명의 직원 정보를 출력하세요.

```python
df.nlargest(10, '연봉(만원)')
```

### 부서를 기준으로 그룹화한 후, 각 그룹 내에서 입사년도가 가장 오래된 직원의 정보를 출력하세요.

```python
df.loc[df.groupby('부서')['입사년도'].idxmin()]
```

### 각 부서에서 연봉 상위 3명의 직원 정보를 출력하세요.

```python
df.groupby('부서').apply(lambda x: x.nlargest(3, '연봉(만원)'))
```

### 연봉의 범위(최소값과 최대값 사이의 차이)가 가장 큰 부서를 찾으세요.

```python
df.groupby('부서')['연봉(만원)'].agg(lambda x: x.max() - x.min()).idxmax()
```
