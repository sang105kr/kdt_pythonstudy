```python
def calculate(x,y) :
  add = x + y
  difference = x - y
  multifly = x * y
  mod = x // y
  remain = x % y
  return (add, difference, multifly, mod, remain)

print(sum(10,20))

print(type(calculate(10,3)))

add, difference, multifly, mod, remain = calculate(10,3)
print(add, difference, multifly, mod, remain )
```

```python
# 1~100까지 정수중에 짝수의 합과 홀수의 합을 각각 구하시오
sum_odd = 0
sum_even = 0;
for num in range(1,100+1):
  if( num % 2 == 1):
    sum_odd += num
  else:
    sum_even += num
print(f'짝수의합={sum_even}, 홀수의합={sum_odd}')
```

```python
def sum_all(*args):
  print(type(args))

  for ele in args:
    print(ele)

sum_all(1,2,3)
```

```python
# 가변인자 *:튜플, *:딕셔너리
def sum_all2(*args1, **args2):
  print(type(args1))
  print(type(args2))

  for ele in args1:
    print(ele)

  for key,val in args2.items():
    print(key, val)

#case1)
sum_all2(1,2,3,name='홍길동',age=10,gender='남자')

#case2)
x = (1,2,3)
y = {'name':'홍길동','age':10,'gender':'남자'}
sum_all2(*x, **y)


```

```python
class Person:
    numberOfPeople = 0

    def __init__(self, name, age):
        self._name = name  # '_' 접두사를 사용하여 필드를 보호(캡슐화)
        self._age = age
        Person.numberOfPeople += 1

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        self._name = value

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        self._age = value

    def smile(self):
        print(f"{self._name} is smiling.")

    def eat(self, food):
        print(f"{self._name} is eating {food}.")

    @staticmethod
    def displayNumberOfPeople():
        print(f"Total number of people: {Person.numberOfPeople}")

p1 = Person('홍길동',30)
print(p1.name)
print(p1._name)
print(p1.name())
```
