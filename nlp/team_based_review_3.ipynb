```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
okt = Okt()
```

```python
def extract_stem(words, tokenizer):
    # ë¬¸ìì—´ì„ í¬í•¨í•œ ë‹¤ì–‘í•œ íƒ€ì…ì—ì„œ NaN í™•ì¸
    if pd.isna(words):
        return

    # ì „ì²´ í…ìŠ¤íŠ¸ì— ëŒ€í•´ í•œ ë²ˆì— í˜•íƒœì†Œ ë¶„ì„ì„ ìˆ˜í–‰
    morphs = tokenizer.morphs(words, stem=True)
    
    # ê²°ê³¼ì—ì„œ ì–´ê°„ë§Œ ì¶”ì¶œí•˜ì—¬ ì¡°ì¸
    return ' '.join(morph for morph in morphs if morph)
```

```python
from tqdm import tqdm
tqdm.pandas()
from functools import partial
```

```python
# extract_stem í•¨ìˆ˜ì— ëŒ€í•œ partial í•¨ìˆ˜ ìƒì„±
extract_stem_okt = partial(extract_stem, tokenizer=okt)
```
