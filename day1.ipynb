### 튜블

```python
a = (1,2,3)
print(type(a))
print(a[0])
print(a[-1])
# 요소변경 안됨
# a[1] = 10
# 요소 삭제 안됨
# del a[1]
# 튜플을 리스트로 변환
b = list(a)
print(type(b))
b[1] = 10
print(b)
# 리스트를 튜플로 변환
c = tuple(b)
print(type(c))
# 요소갯수 확인
print(len(c))
```
