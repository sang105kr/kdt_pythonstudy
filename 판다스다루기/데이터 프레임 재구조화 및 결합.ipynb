### merge 예제

```python
# key 컬럼명이 같은경우
left = pd.DataFrame({
    'KEY':['K0','K1','k2','K3'],
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3']
})
right = pd.DataFrame({ㅁ
    'KEY':['K0','K1','K3','K4'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']    
})
display(left)
display(right)
```

```python
# merge 기본은 inner join
result = pd.merge(left,right,on='KEY')
display(result)

# left outer join
result = pd.merge(left,right,on='KEY',how='left')
display(result)

# right outer join
result = pd.merge(left,right,on='KEY',how='right')
display(result)

# full outer join
result = pd.merge(left,right,on='KEY',how='outer')
display(result)
```

```python
# key 컬럼명이 다른경우
left = pd.DataFrame({
    'KEY_LEFT':['K0','K1','k2','K3'],
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3']
})
right = pd.DataFrame({
    'KEY_RIGHT':['K0','K1','K3','K4'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']    
})
display(left)
display(right)
```

```python
# merge 기본은 inner join
result = pd.merge(left,right,left_on='KEY_LEFT', right_on='KEY_RIGHT',)
display(result)

# left outer join
result = pd.merge(left,right,left_on='KEY_LEFT', right_on='KEY_RIGHT',how='left')
display(result)

# right outer join
result = pd.merge(left,right,left_on='KEY_LEFT', right_on='KEY_RIGHT',how='right')
display(result)

# full outer join
result = pd.merge(left,right,left_on='KEY_LEFT', right_on='KEY_RIGHT',how='outer')
display(result)
```

```python
# merge 기본은 inner join
result = left.merge(right,left_on='KEY_LEFT', right_on='KEY_RIGHT',)
display(result)

# left outer join
result = left.merge(right,left_on='KEY_LEFT', right_on='KEY_RIGHT',how='left')
display(result)

# right outer join
result = left.merge(right,left_on='KEY_LEFT', right_on='KEY_RIGHT',how='right')
display(result)

# full outer join
result = left.merge(right,left_on='KEY_LEFT', right_on='KEY_RIGHT',how='outer')
display(result)
```

### join() 예제

```python
# key 컬럼명이 같은경우
left = pd.DataFrame({
        'A':['A0','A1','A2','A3'],
        'B':['B0','B1','B2','B3']
    },
    index=['K0','K1','k2','K3'])
right = pd.DataFrame({
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']    
    },
    index=['K0','K1','K3','K4'])

display(left)
display(right)
```

```python
# join 기본은 left outer join
display(left.join(right))

# inner join
display(left.join(right,how='inner'))

# right outer join
display(left.join(right,how='right'))

# full outer join
display(left.join(right,how='outer'))
```
