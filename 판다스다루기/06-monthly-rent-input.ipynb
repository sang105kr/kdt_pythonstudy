## 데이터셋 불러오기

````python
# ./data/monthly_2021/*.csv
# glob 로 목록 출력하기
import glob
paths = glob.glob('./data/*.csv')
paths
````

````python
# columns={"이동거리(M)": "이용거리(M)", "이용시간(본)":"이용시간(분)"} 컬럼명 일관되게 변경하기
# temp 에 데이터 프레임 담기
temp = []
for pth in paths:
    df_temp = pd.read_csv(pth, encoding='cp949')
    df_temp = df_temp.rename(columns={'이동거리(M)':'이용거리(M)', '이용시간(본)':'이용시간(분)'})
    temp.append(df_temp)
temp  


````

## 데이터 타입 변경

````python
# 운동량, 탄소량 to_numeric 으로 변경하기
# to_numeric() : 
#    1. 자동 타입변환 : 'int', 'float'중 가장 적합한 숫자 타입으로 결정한다.
#    2. 오류처리 : 'errors' 매개변수를 통해 변환중 발생될 수 있는 오류를 유연하게 처리해준다.
#                  errors='coerce' 옵션을 사용하면 변경할수 없는 값은 NaN(Not a Number)로 변환된다.
#    3. 유연한 데이터 처리 : 숫자가 문자열등으로 잘못 입력되었을 경우 적절히 숫자형으로 변환
raw[['운동량','탄소량']].head(5)
raw["운동량"] = pd.to_numeric(raw['운동량'], errors='coerce')
raw["탄소량"] = pd.to_numeric(raw['탄소량'], errors='coerce')
raw[['운동량','탄소량']].head(5)
````

````python
df['연령대코드'].value_counts()
````

````python
# 문자열이 여러줄로 되어 있을때  ''' or """" 로 감싼다
age_code = """
10대 : AGE_001
20대 : AGE_002
30대 : AGE_003
40대 : AGE_004
50대 : AGE_005
60대 : AGE_006
70대 : AGE_007
기타 : AGE_008
"""
age_code
````

````python
# age_dict 값 만들기
age_dict = {}
for age in age_code.split('\n')[1:] :
    print(age)
    if ":" in age :
        key = age.split(":")[1].strip()
        val = age.split(":")[0].strip()
        age_dict[key] = val
print(age_dict)
````

### map 활용하기

````python
# "연령대" 파생변수 만들기
# df['연령대'] = df["연령대코드"].map(lambda key : age_dict[key])
df['연령대'] = df["연령대코드"].map(age_dict)
df.head(1) 
````

````python
# 연령대 빈도수 구하기
# values 정렬
# age_count = df['연령대'].value_counts().sort_values()
# index 정렬
age_count = df['연령대'].value_counts().sort_index()
age_count
````

## 연령대 vs 성별

````python
# index="연령대", columns="성별", values="이용건수" 합계
age_gender = pd.pivot_table(data=df, index="연령대", columns="성별", values="이용건수",aggfunc=np.mean)
age_gender
````

````python
age_gender = age_gender.loc[ : ,['F','M']]
````

````python
age_gender_rate = age_gender.loc[ : ,['F','M']]
age_gender_rate
````

````python
# 세로방향 합계 axix = 0
sum_of_col = age_gender.loc[ : ,['F','M']].sum()
# 가로방향 합계 axix = 1
sum_of_row = age_gender.loc[ : ,['F','M']].sum(axis=1)

sum_of_col,sum_of_row
````

````python
# 세로방향 합계 axis = 0
sum_of_col = age_gender.loc[ : ,['F','M']].sum()
age_gender.loc['합계'] = sum_of_col

# 가로방향 합계 axix = 1
sum_of_row = age_gender.loc[ : ,['F','M']].sum(axis=1)

age_gender['합계'] = sum_of_row
age_gender
````

````python
# 연령대, 성별, 이용건수 시각화
age_gender.iloc[ :-2, : -1].plot.bar(rot=30)
````

````python
age_gender_rate
````

````python
# age_gender_rate['F_R'] = age_gender_rate.sum()
print(age_gender_rate.sum()[0],age_gender_rate.sum()[1])
print(age_gender_rate.sum().sum())

print(age_gender_rate.sum()[0]/age_gender_rate.sum().sum() *100)
print(age_gender_rate.sum()[1]/age_gender_rate.sum().sum() *100)

f_r = age_gender_rate.sum()[0]/age_gender_rate.sum().sum() * 100
m_r = age_gender_rate.sum()[1]/age_gender_rate.sum().sum() * 100

age_gender_rate.loc['비율'] = [f_r, m_r]
age_gender_rate
````

````python
age_gender_rate['F_R'] =   age_gender_rate['F'] / (age_gender_rate['F'] + age_gender_rate['M']) * 100
age_gender_rate['M_R'] =   age_gender_rate['M'] / (age_gender_rate['F'] + age_gender_rate['M']) * 100
age_gender_rate
````

````python
# 성별 차이가 많이 나는 연령대는?
np.abs(age_gender_rate['F_R']-age_gender_rate['M_R']).sort_values(ascending=False).head(1)
````
